```python
#Demonstração gráfica da relçao dos serviços com o Churn
#Graphic demonstration of the relationship between services and Churn

colunas = ['PhoneService', 'MultipleLines', 'Internet', 'OnlineSecurity',
            'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'LongTerm_Contract', 'PaperlessBilling', 'AutomaticPayment']

 
fig, axes = plt.subplots(3, 4, figsize=(14, 8), sharey=True)

# Iterando sobre as colunas
for i, coluna in enumerate(colunas):
    # Filtrando os dados para pessoas com o serviço (valor 1) e calculando a proporção para Churn
    coluna_serv = df_clientes_analise.loc[df_clientes_analise[coluna] == 1, 'Churn'].value_counts(normalize=True).sort_index() * 100
    
    # Criando um gráfico de barra
    sns.barplot(x=coluna_serv.index, y=coluna_serv.values, ax=axes[i//4, i%4], palette="viridis")
    
    # Configurando rótulos e título
    axes[i//4, i%4].set_title(coluna)
    axes[i//4, i%4].set_xlabel("Churn")
    axes[i//4, i%4].set_ylabel("Percentagem")
#i // 4 representa a linha do subplot.
#i % 4 representa a coluna do subplot.
#Essas operações são usadas para converter um índice único (i) em coordenadas de linha e coluna em uma grade 3x4.
# Ajustando o layout
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

## Considerações
Observe que os serviços de Online Security, Tech Suport, Long Term Contract e Automatic Paymente possuem uma proporção de Churn menor. Já Streamings, Internet e Paperless Billing possuem um Churn maior #analisar por que esses serviços não são mais atrativos para o consumidor

Note that Online Security, Tech Support, Long Term Contract and Automatic Payment services have a lower Churn ratio. Streaming, Internet and Paperless Billing have a higher Churn #analyze why these services are no longer attractive to consumers

Agora eu quero saber se contratar mais serviços influencia no Churn. Para isso, primeiro eu preciso criar uma coluna com a soma dos serviços

Now I want to know if adding more services influences churn. For this, I first need to create a column with the sum of services.


```python
# Lista das colunas que representam os serviços
colunas_servicos = ['PhoneService', 'MultipleLines', 'Internet', 'OnlineSecurity',
                    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
                    'StreamingMovies', 'LongTerm_Contract', 'PaperlessBilling', 'AutomaticPayment']

# Adiciona uma nova coluna representando o número total de serviços contratados por cliente
df_clientes_analise['TotalServicos'] = df_clientes_analise[colunas_servicos].sum(axis=1)
df_clientes_analise.head()
```

```python
df_clientes_analise['TotalServicos'].value_counts(normalize = True)*100
```

```python
# Analisa se a assinatura de mais serviços influencia no churn
servicos_churn  = pd.DataFrame
servicos_churn = df_clientes_analise.groupby(['TotalServicos','Churn'])['Churn'].count()
servicos_churn
```

```python
total_clientes_ativos = servicos_churn.groupby('TotalServicos').sum()
total_clientes_ativos.sum()
```

```python
taxa_churn = (servicos_churn / total_clientes_ativos) * 100

# Exiba os resultados
print(taxa_churn)
```

```python
df_clientes_mult_servicos = df_clientes_analise.loc[df_clientes_analise['TotalServicos'] > 7,['Churn']].value_counts (normalize = True).sort_index()*100
df_clientes_mult_servicos
```

```python
df_clientes_menos_servicos = df_clientes_analise.loc[df_clientes_analise['TotalServicos'] <= 7,['Churn']].value_counts (normalize = True).sort_index()*100
df_clientes_menos_servicos
```
